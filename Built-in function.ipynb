```python
""" divide the number 148.789 by 12,13,14 and 15 separtely, and their quotient ,remainder add"""
a=divmod(148.789,12)
b=divmod(148.789,13)
c=divmod(148.789,14)
d=divmod(148.789,15)
print(a,b,c,d)
q=a[0]+b[0]+c[0]+d[0]
r=a[1]+b[1]+c[1]+d[1]
print("Sum of quotients:",round(q))
print("Sum of remainder:",round(r))
```

```python
""" Ask the user to enter 5 values and find the minimum and maximum"""
a=int(input("1st value"))
b=int(input("2nd value"))
c=int(input("3rd value"))
d=int(input("4th value"))
e=int(input("5th value"))
print(a,b,c,d,e)
print("Maximum value:",max(a,b,c,d,e))
print("Minimum value:",min(a,b,c,d,e))
```
