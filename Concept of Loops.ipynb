# Continue statement

```python
i=1
while True:
    if i%9!=0:
        print("Inside if")
        i+=1
```

```python
i = 1
while True:                 # infinite loop start
    if i % 9 != 0:          # agar i 9 se divisible nahi hai
        print("Inside if")
        i += 1              # i ko 1 se badhao
        continue            # fir se loop shuru
    print("Something")      # jab i % 9 == 0 hoga tab yaha aayega
    print("Something else")
    break                   # loop tod do
print("Done")
```

```python
""" Write a program to print a series of 22, 222,2222,22222,222222 till the value given by user """
n=int(input("Enter the vlue"))
i=1
p=2
while i<=n:
    p=(p*10)+2
    print(p)
    i+=1
    continue
    print("this is the series")
    break
print("Done")
```

# For Loop :-

```python
""" Write a program and to print the square of number till 10 """
L=[]
for i in range(10):
    print(i+1)
    L.append(i**2)
print(L)
```

```python
L=[]
for i in range(0,10,2):
    print(i+1)
    L.append(i**2)
print(L)
```

```python
""" Write a  program to separate positve and negative integers from the list"""
L=[24,-15,89,12,36,-85,0,-2,-9,0,-31,32,14]
L1=[]
L2=[]
L3=[]
for i in range(13):
    if L[i]>0:
        L1.append(L[i])
    elif L[i]<0:
        L2.append(L[i])
    else:
        L3.append(L[i])
print("The list of positive number are",L1)
print("The list of negative number are",L2)
print("The list of Zero number are",L3)
```

```python
""" Write a program to separate the even and odd in the list"""
L=[82,64,27,39,74,93,28,84,45,37]
len(L)
L1=[]
L2=[]
for i in range(10):
    if L[i]%2==0:
        L1.append(L[i])
    else:
        L2.append(L[i])
print("The list of Even number are",L1)
print("The list of Odd number are",L2)
        
```
