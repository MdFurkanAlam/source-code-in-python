# Accessing dict elements

```python
d1={102:'Rahul',105:'Payal',106:'Arjun',107:'Prachi'}
print(d1)
```

```python
d1={102:'Rahul',105:'Payal',106:'Arjun',107:'Prachi'}
print(d1[102],d1[105],d1[106],d1[107])
```

```python
d1={102:'Rahul',105:'Payal',106:'Arjun',107:'Prachi'}
for k in d1:
    print(k) # only keys
```

```python
d1={102:'Rahul',105:'Payal',106:'Arjun',107:'Prachi'}
for k in d1:
    print(k,d1[k]) 
```

# How to edit dect element and add new element

```python
d1={102:'Rahul',105:'Payal',106:'Arjun',107:'Prachi'}
d1[102]='Furkan' #because 102 key is exist in d1 then change data value
print(d1)
```

```python
d1={102:'Rahul',105:'Payal',106:'Arjun',107:'Prachi'}
d1[108]='Furkan' #because  key is not exist in d1 then create a new key and data value
print(d1)
```

# method

```python
d={1:'Furkan',2:'Sameer',3:'Rashid',4:'Arif'}
d.keys()
```

```python
d={1:'Furkan',2:'Sameer',3:'Rashid',4:'Arif'}
d.values()
```

```python
d={1:'Furkan',2:'Sameer',3:'Rashid',4:'Arif'}
d.items()
```

```python
d={1:'Furkan',2:'Sameer',3:'Rashid',4:'Arif'}
for t in d.items():
    k=t[0]
    d=t[1]
    print(k,d)
```

```python
d={1:'Furkan',2:'Sameer',3:'Rashid',4:'Arif'}
for t in d.items():
    k,d=t
    print(k,d)
```

```python
d={1:'Furkan',2:'Sameer',3:'Rashid',4:'Arif'}
for k,d in d.items(): # k and d is tuple
    print(k,d)
```
