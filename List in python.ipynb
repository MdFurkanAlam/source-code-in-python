# clear() method in list

```python
a=[2,9,92,84,84,75] 
b=[6.4,7.8,9.4,2.3]
a.append(b)
print(a)
print(len(a))


a.clear() # clear all items in form the list
print(a)
print(len(a))


a.append(b)
print(a)
print(len(a))
```

```python
""" take the input of the list from the user and then reverse the list """
L=[]
for i in range(5):
    L.append(int(input("Enter Elements: ")))
print(L)
L.reverse()
print(L)
```

```python
""" take the input from the user and interchange the first and the last element of the list"""
L=[]
n= int(input("How many items you want in the list "))
for i in range(n):
    L.append(int(input("Enter Elements: ")))
print(L)
temp = L[0]
L[0]= L[n-1]
L[n-1]=temp
print(L)



```
