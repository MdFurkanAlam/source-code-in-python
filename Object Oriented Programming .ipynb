```python
class bank_account:
    def __init__(self,balance):
        self.__balance=balance

    def deposit(self,amount):
        self.__balance=self.__balance+amount

    def withdraw(self,amount):
        if self.__balance>=amount:
            self.__balance=self.__balance-amount
        else: 
            return False
    def get_balance(self):
        return self.__balance
```

```python
furkan=bank_account(4733)
furkan.get_balance()
```

```python
furkan.deposit(737)
furkan.get_balance()
```

```python
furkan.withdraw(100)
furkan.get_balance()
```

# abstraction

```python
import abc
class ihh:
    @abc.abstractmethod
    def student_details(self):
        pass

    @abc.abstractmethod
    def homework_details(self):
        pass

    @abc.abstractmethod
    def exam_details(self):
        pass
```

```python
class demo(ihh):
    def student_details(self):
        print('student are available here')
    def homework_details(self):
        print(3+4)
```

```python
test_obj=test()
test_obj.student_details()
test_obj.exam_details()
```

# Decorators

```python
def check():
    print('this is starting of the function')
    print('this is check function')
    print(2+3)
    print('this is ending of the function')

check()
```

```python
def decor(func):
    def inner_decor():
        print('this is starting of the function')
        func()
        print('this is ending of the function')
    return inner_decor()

@decor
def check1():
    print(2+5)
```

```python
@decor
def learn():
    print(3730984007.5*2)
```

```python
import time
def timer_test(func):
    def inner_time_test():
        start=time.time()
        print("Start",start)
        func()
        end=time.time()
        print('End',end)
        print(end-start)
    return inner_time_test()
@timer_test
def calculate():
    print(5+5)
```

# property decorators- getters,setters,and delete

```python
class movie:
    def __init__(self,movie_name,ticket_price):
        self.movie_name=movie_name
        self.__ticket_price=ticket_price

    @property
    def ticket_price_access(self):
        return self.__ticket_price

    @ticket_price_access.setter
    def ticket_price_set(self,price):
        if price<=500:
            pass
        else:
            self.__ticket_price=price
```

```python
mov=movie('Avenger','500')
mov.ticket_price_access
```

```python
mov.ticket_price_set=700
```

```python
mov.ticket_price_access
```

```python
class movie:
    def __init__(self,movie_name,ticket_price):
        self.movie_name=movie_name
        self.__ticket_price=ticket_price

    @property
    def ticket_price_access(self):
        return self.__ticket_price

    @ticket_price_access.setter
    def ticket_price_set(self,price):
        if price<=500:
            pass
        else:
            self.__ticket_price=price

    @ticket_price_access.deleter
    def ticket_price_delete(self):
        del self.__ticket_price
```

```python
mov1=movie('Iron Man',400)
```

```python
mov1.ticket_price_access
```

```python
mov1.ticket_price_set=600
```

```python
mov1.ticket_price_access
```

```python
del mov1.ticket_price_delete
```

```python
mov1.ticket_price_access
```
