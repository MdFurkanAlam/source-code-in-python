intersection() set object method

```python
set1 = {1, 2, 3, 4, 5}
set2 = {3, 4, 5, 6, 7}
set3 = {5, 7, 8, 9}

# Finding the intersection of set1 and set2
intersection_result1 = set1.intersection(set2)
print(f"Intersection of set1 and set2: {intersection_result1}")

# Finding the intersection of set1, set2, and set3
intersection_result2 = set1.intersection(set2, set3)
print(f"Intersection of set1, set2, and set3: {intersection_result2}")
```

unions() set object method

```python
set1 = {1, 2, 3}
set2 = {3, 4, 5}
set3 = {5, 6, 7}

# Union of two sets
new_set = set1.union(set2)
print(new_set)  # Output: {1, 2, 3, 4, 5}

# Union of multiple sets
another_new_set = set1.union(set2, set3)
print(another_new_set) # Output: {1, 2, 3, 4, 5, 6, 7}
```

issubset() set object method

```python
# Define two sets
set_a = {1, 2, 3}
set_b = {1, 2, 3, 4, 5}
set_c = {1, 6}

# Check if set_a is a subset of set_b
is_subset_ab = set_a.issubset(set_b)
print(f"Is set_a a subset of set_b? {is_subset_ab}")  # Output: True

# Check if set_c is a subset of set_b
is_subset_cb = set_c.issubset(set_b)
print(f"Is set_c a subset of set_b? {is_subset_cb}")  # Output: False
```

issuperset() set object method

```python
set_A = {1, 2, 3, 4, 5}
set_B = {2, 4}
set_C = {1, 6}

print(set_A.issuperset(set_B)) # Output: True (because all elements of set_B are in set_A)
print(set_A.issuperset(set_C)) # Output: False (because 6 is in set_C but not in set_A)
print(set_B.issuperset(set_A)) # Output: False (because set_B does not contain all elements of set_A)
```
