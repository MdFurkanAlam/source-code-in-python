# Reading and wrting files

```python
data={
    "name":"Furkan",
    "email":"furkan@gmail.com",
    "phone number": 7461968015,
    "subject":["data science" ,"big data","data analytics"]
}
```

```python
import json
```

```python
with open("data.json","w") as f:
    json.dump(data,f)
```

```python
import csv
```

```python
data =[
    ["name ","email","phone number"],
    ["Furkan","furkan@gmail.come",7461968015],
    ["Rashid","rashid@gmail.com",70704156994],
    ["Arif", "arif@gmail.com",9199916628]
]
```

```python
with open("data.csv","w") as f:
    writer=csv.writer(f)
    for i in data:
        writer.writerow(i)
```

```python
with open("data.csv","r") as f:
    read_data=csv.reader(f)

    for i in read_data:
        print(i)
```

```python
with open("data.bin","wb") as f:
    f.write(b"\x01\x02\x03\x3454235")
```

```python
with open("data.bin","rb") as f:
    print(f.read())
```

# Buffered Read And write othe files methods

```python
import io
```

```python
with open("data1.txt","wb") as f:
    file = io.BufferedWriter(f)

    file.write(b"In the context of file management and computer command lines, pwd is an abbreviation for Print Working Directory or Present Working Directory. The pwd command is a standard utility in Unix-like operating systems (such as Linux and macOS) used to display the full, absolute path of the directory you are currently working in. This helps users navigate the file system and confirm their exact location. \n")
    file.write(b"this is my first line\n")
    file.write(b"this is my second line\n")
    file.flush()
```

```python
with open("data1.txt","rb") as f:
    file= io.BufferedReader(f)

    print(file.read(400))
```
