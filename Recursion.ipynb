# Recursion in python

1. write a recurssive function to calculate sum of first n natural nuumber

```python
def sum(n):
    if n==1:
        return 1
    s=n+sum(n-1)
    return s

x=sum(4)
print(x)
```

1. write a recurssive function to calculate sum of squares of first n natural nuumber

```python
def sumsquare(n):
    if n==1:
        return 1
    s=n*n+sumsquare(n-1)
    return s

x=sumsquare(4)
print(x)
```
