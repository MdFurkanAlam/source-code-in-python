# using loops in tuples

```python
""" separate positive and negative integers from the tuple """
a=(74,65,-34,-75,75,-54,-54,33,53)
l1=[]
l2=[]
for item in a:
    if item>=0:
        l1.append(item)
    else:
        l2.append(item)
print(tuple(l1))
print(tuple(l2))
```

```python
""" multiply adjacent elementss in tuple """
a=(8,3,5,3,6,7,5)
L=list(a)
b=[]
for i in range(len(L)-1):
    b.append(L[i]*L[i+1])
    i+=1
print(tuple(b))
```

```python
""" orignal list of tuple [(1,2),(2,3),(3,4)]
convert it into list of lidts [[1,2],[2,3],[3,4]]"""
T=[(1,2),(2,3),(3,4)]
L=[]
for i in range(len(T)):
    L.append(list(T[i]))
print(L)
```

```python
""" original list of tuple [(1,2),(2,3),(3,4)]
convert [3,5,7] """
T=[(1,2),(2,3),(3,4)]
L=[]
L1=[]
for i in range(len(T)):
    L.append(list(T[i]))
    for j in range(len(T[i])-1):
        L1.append(T[i][j]+T[i][j+1])
print(L1)
```

```python
""" original list of tuple [(1,2),(2,3),(3,4)]
convert [3,5,7] """
T = [(1,2), (2,3), (3,4)]
L1 = []

for i in range(len(T)):
    L1.append(T[i][0] + T[i][1])

print(L1)
```

```python
""" original list of tuple [(1,2),(2,3),(3,4)]
convert [3,5,7] """
T = [(1,2), (2,3), (3,4)]
L1 = [x + y for (x, y) in T]
print(L1)
```
