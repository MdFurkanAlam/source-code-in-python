# *args in python

```python
def argfunc(*args):
    s=0
    for i in range(len(args)):
        s+=args[i]
    print(s)

argfunc(3,5,2,5,24,34,23,43.3,4,43,432.3,-33.3,-3.43,-64.4,34,3.3,656,54)
```

# Generator function

```python
def test_fib(n):
    a=0
    b=1
    for i in range(n):
        yield a
        a,b=b,a+b

test_fib(10)
```

```python
for i in test_fib(10):
    print(i)
```

Calculate factorial of a number using recursive lambda expresion

```python
def fact(n):
    if n==0 or n==1:
        return 1
    return n+fact(n-1)
fact(10)
```

```python
f=lambda n: 1 if n==0 or n==1 else n+f(n-1) #using recursive  lambda expresion
f(3)
```
